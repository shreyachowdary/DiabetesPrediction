# Exploratory Data Analysis - Pima Indians Diabetes Dataset

This notebook explores the dataset used for diabetes prediction.

```python
import pandas as pd
from pathlib import Path

data_path = Path("../data/diabetes.csv")
df = pd.read_csv(data_path)
df.head(10)
```

```python
df.info()
df.describe()
```

```python
# Check for zeros (missing value indicators in Pima dataset)
zero_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for col in zero_cols:
    print(f"{col}: {(df[col] == 0).sum()} zeros")
```
